```python
import pandas as pd
```

```python
df=pd.read_csv("TBESH_formatted_plural.csv",header=None,sep="\t")
df.columns=["no","H","morph","gloss","meaning","len"]

df0=pd.read_csv("TBESH_formatted.csv",header=None,sep="\t")
#df0.columns=["no","H","morph","gloss","meaning","len"]
```

```python
def repl(l):
    return l.replace("(a)","|").replace("(b)","|").replace("(c)","|").replace("(d)","|").replace("(e)","|").replace("(f)","|")
```

```python
def repl2(l):
    return [x.strip() for x in l.split("|")][1:]

def uniq(l):
    if len(set(l))==1: return l[0]
    else: return "|".join(l)
```

```python
df2=df.copy()
df2["morph"]=df2["morph"].apply(repl)
df2["gloss"]=df2["gloss"].apply(repl)
df2["meaning"]=df2["meaning"].apply(repl)
df2
df2["morph_list"]=df2["morph"].apply(repl2).apply(uniq)
df2["gloss_list"]=df2["gloss"].apply(repl2).apply(uniq)
df2["meaning_list"]=df2["meaning"].apply(repl2).apply(uniq)
df2
```

```python
df2.to_csv("TBESH_formatted_2.csv",sep="\t",index=False)
```
