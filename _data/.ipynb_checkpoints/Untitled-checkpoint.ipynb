```python
df["24t"]=df["24"].apply(lambda x: x.replace("kh","x").replace("ph","f").replace("th","þ") if not pd.isnull(x) else None)
```

```python
list1=["ā→a","ă→a","â→a","î→i","ǝ→e","š→sh","ḥ→ch","ô→o","ŏ→o","û→u","ʿ→'","ē→e","ĕ→e","ʾ→'","ṣ→ts","ō→o","ś→s","ṭ→t"]
subst={x.split("→")[0]:x.split("→")[1] for x in list1}
```

```python
def rep(l):
    if pd.isnull(l): return None
    if l[0]=="ʾ": return rep(l[1:])
    if l[-1]=="ʾ": return rep(l[:-1])
    if l[0]=="ʿ": return rep(l[1:])
    if l[-1]=="ʿ": return rep(l[:-1])

    for key in  subst.keys():
        l=l.replace(key,subst[key])
    return l
```

```python
df["22t"]=df["22"].apply(rep)
pd.Series(df[["22","22t","24t"]].apply(lambda x: None if pd.isnull(x["22"]) else [x["22t"][i1:i2]+"→"+x["24t"][j1:j2] for tag, i1, i2, j1, j2 in difflib.SequenceMatcher(None, x["22t"], x["24t"]).get_opcodes() if x["22t"][i1:i2]!=x["24t"][j1:j2]] ,axis=1).dropna().sum()).value_counts()
```

```python
df["22t-24t-diff"]=df[["22t","24t"]].apply(lambda x: None if pd.isnull(x["22t"]) else [x["22t"][i1:i2]+"→"+x["24t"][j1:j2] for tag, i1, i2, j1, j2 in difflib.SequenceMatcher(None, x["22t"], x["24t"]).get_opcodes() if x["22t"][i1:i2]!=x["24t"][j1:j2]] ,axis=1)
df[["22","24","22t-24t-diff"]].dropna()[df["22t-24t-diff"].dropna().apply(len)>0]
```
